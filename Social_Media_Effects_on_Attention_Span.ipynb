```python
# prompt: convert string data type to numerical data type by assigning numerical values. take column names from the file provided

from google.colab import drive
import pandas as pd

drive.mount('/content/drive')


# Replace 'your_file_path' with the actual path to your file in Google Drive
df = pd.read_csv('/content/drive/MyDrive/data.csv')

# Get column names from the DataFrame
column_names = df.columns.tolist()

# Create a dictionary to map string values to numerical values
string_to_numerical = {}
numerical_counter = 0

# Iterate through each column and assign numerical values to unique string values
for column in column_names:
  unique_values = df[column].unique().tolist()
  for value in unique_values:
    if value not in string_to_numerical:
      string_to_numerical[value] = numerical_counter
      numerical_counter += 1


# Convert the string values to numerical values in the DataFrame
for column in column_names:
  df[column] = df[column].map(string_to_numerical)


print(df.head(20)) # Display the first few rows of the updated DataFrame
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
target_columns = [
    'Age_Group', 'Gender', 'Education_Level', 'Occupation',
    'Average_Screen_Time', 'Device', 'Screen_Activity', 'App_Category',
    'Screen_Time_Period', 'Environment', 'Productivity', 'Attention_Span',
    'Work_Strategy', 'Notification_Handling', 'Usage_of_Productivity_Apps'
]

for col in target_columns:
    print('\
Value counts for', col)
    print(df[col].value_counts())
    fig, ax = plt.subplots(figsize=(12, 8))
    sns.countplot(data=df, x=col, color='#2563EB', ax=ax)
    ax.set_title('Distribution of ' + col.replace('_', ' '), fontsize=20, pad=15, color='#171717', fontweight='semibold')
    ax.set_xlabel(col.replace('_', ' '), fontsize=16, labelpad=10, color='#171717', fontweight='medium')
    ax.set_ylabel('Count', fontsize=16, labelpad=10, color='#171717', fontweight='medium')
    ax.tick_params(axis='x', labelsize=14, colors='#171717', rotation=45)
    ax.tick_params(axis='y', labelsize=14, colors='#171717')
    ax.set_axisbelow(True)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color('#171717')
    ax.spines['bottom'].set_color('#171717')
    plt.grid(axis='y', color='#F3F4F6')
    plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.2)
    plt.tight_layout()
    plt.show()

print('Displayed value counts and countplots for all main categorical columns.')
```

```python
# Create correlation heatmap for numeric columns
import numpy as np

# Convert categorical columns to numeric
df_numeric = df.copy()
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df_numeric[col] = pd.Categorical(df_numeric[col]).codes

# Create correlation matrix
corr_matrix = df_numeric.corr()

# Plot correlation heatmap
fig, ax = plt.subplots(figsize=(12, 8))
sns.heatmap(corr_matrix,
            annot=True,
            cmap='RdYlBu',
            center=0,
            ax=ax)

ax.set_title('Correlation Heatmap of Variables',
             fontsize=20,
             pad=15,
             color='#171717',
             fontweight='semibold')

plt.tight_layout()
plt.show()

# Create donut charts for key categorical variables
def create_donut_chart(data, column, title):
    counts = data[column].value_counts()
    fig, ax = plt.subplots(figsize=(12, 8))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', pctdistance=0.85,
            colors=['#2563EB', '#24EB84', '#B2EB24', '#EB3424', '#D324EB'])
    centre_circle = plt.Circle((0,0), 0.70, fc='white')
    fig.gca().add_artist(centre_circle)
    plt.title(title, fontsize=20, pad=15, color='#171717', fontweight='semibold')
    plt.show()

# Create donut charts for key variables
key_variables = ['Gender', 'Education_Level', 'Productivity', 'Device']
for var in key_variables:
    create_donut_chart(df, var, f'Distribution of {var.replace("_", " ")}')

# Create line chart for screen time patterns
screen_time_counts = df.groupby(['Screen_Time_Period', 'Average_Screen_Time']).size().unstack()
screen_time_counts = screen_time_counts.fillna(0)

fig, ax = plt.subplots(figsize=(12, 8))
for col in screen_time_counts.columns:
    ax.plot(screen_time_counts.index, screen_time_counts[col], marker='o', linewidth=2, label=col)

ax.set_title('Screen Time Patterns Throughout the Day',
             fontsize=20,
             pad=15,
             color='#171717',
             fontweight='semibold')
ax.set_xlabel('Time Period', fontsize=16, labelpad=10, color='#171717', fontweight='medium')
ax.set_ylabel('Number of Users', fontsize=16, labelpad=10, color='#171717', fontweight='medium')
ax.tick_params(axis='both', labelsize=14, colors='#171717')
ax.legend(title='Average Screen Time (hours)', bbox_to_anchor=(1.05, 1), loc='upper left')
ax.grid(True, axis='y', color='#F3F4F6')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
